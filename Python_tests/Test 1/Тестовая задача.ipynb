```python
import numpy as np
import matplotlib.pyplot as plt
```

# Исходные данные

```python
C = 1.0
U = 1.0
L = 1.0

t_k = 100
t_0 = 0
```

# Системма уравнений

```python
def sys(t, X):
    I = X[0]
    q = X[1]
    
    dI_dt = - q / (L * C)
    dq_dt = I
    
    return np.array([dI_dt, dq_dt])
```

# Метод Ронге-Кутты

```python
def rk4(init, sys, t0, dt, n):

        t = np.zeros(n)
        m = len(init) + 1
        res = np.zeros((n, m))

        i = 0
        X = init
        t[i] = t0
        res[i, 0] = t[i]
        res[i, 1:m] = X


        while i < n - 1:
            k1 = sys(t[i], X)
            k2 = sys(t[i] + 0.5*dt, X + k1*0.5*dt)
            k3 = sys(t[i] + 0.5*dt, X + k2*0.5*dt)
            k4 = sys(t[i] + dt, X + k3*dt)
            i += 1
            X += (k1 + 2*k2 + 2*k3 + k4)*dt/6
            t[i] = t[i - 1] + dt
            res[i, 0] = t[i]
            res[i, 1:m] = X

        return res[0:i + 1, :]
```

# Решение уравнения

```python
init = np.array([0, U * C])
n = 10000
dt = (t_k - t_0) / n

res = rk4(init, sys, 0, dt, n)
t = res[:, 0]
I = res[:, 1]
q = res[:, 2]
```

## Напряжение на конденсаторе

```python
u = q / C
```

# Построение графиков

```python
%config InlineBackend.figure_format = 'svg'

plt.plot(t, I)
plt.xlabel("$t$, с")
plt.ylabel("$I$, А")
plt.grid()
plt.show()

plt.plot(t, u)
plt.xlabel("$t$, с")
plt.ylabel("$u$, В")
plt.grid()
plt.show()
```
